```python
sent = '''The fiber is most often spun into yarn or thread 
and used to make a soft, rigid, and flexible textile. '''
```

```python
# Find the synonyms of all the adjective in the sentence
```

```python
from nltk import pos_tag
from nltk.tokenize import word_tokenize
```

```python
adj = [word[0] for word in pos_tag(word_tokenize(sent))
      if word[1].startswith('JJ')]
```

```python
for word in adj:
    syns = set()
    for syn in wordnet.synsets(word):
        syns.add(syn.lemmas()[0].name())
    syns.discard(word)
    print('Synonyms of',word,'are:',syns)
```

#### antonyms or opposite words

```python
ant = set()
for synset in wordnet.synsets("pass"):
    for lemma in synset.lemmas():
        if lemma.antonyms():    #When antonyms are available, add them into the list
            ant.add(lemma.antonyms()[0].name())
print('Antonyms: ', ant)
```
