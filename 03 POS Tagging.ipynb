```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
sents = '''English is a West Germanic language in the Indo-European language family, whose speakers, called Anglophones, originated in early medieval England.[4][5][6] The namesake of the language is the Angles, one of the ancient Germanic peoples that migrated to the island of Great Britain. Modern English is both the most spoken language in the world[7] and the third-most spoken native language, after Mandarin Chinese and Spanish.[8] It is also the most widely learned second language in the world, with more second-language speakers than native speakers.'''
```

```python
word_tokenize(sents)
```

```python
tags = pos_tag(word_tokenize(sents))
```

```python
tags
```

```python
# Extract verbs
for word, tag in tags:
    if tag.startswith('V') and word.isalpha():
        print(word)
```

```python
# How many proper nouns are present?
count = 0
for word, tag in tags:
    if tag == 'NNP' and word.isalpha():
        count += 1
count
```

```python
# Extract all the pairs of adjective and noun
for i in range(len(tags)):
    if tags[i][1].startswith('J') and tags[i+1][1].startswith('N'):
        print(tags[i][0], tags[i+1][0])
```
